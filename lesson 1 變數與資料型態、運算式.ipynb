### 1.1變數

  * 1.1.1 物件
  
  在python當中，有著最基本的內建資料型態
  分別為:
  
  1.整數(integer) 就是整數 ex:123 
  
  2.浮點數(float) 就是有小數點的數字ex:1.23，或是有指數的東西，ex:1.0e5，他代表10的5次方也就是100000.0 
  
  3.布林值(boolean) 就是 是否 ex:True
  
  4.字串(String) 就是用單(雙)引號包起來的東西 ex: "你好"、'我是'、"'字串'"、"123"、'1.23'、'True' 
  

  
  
  * 1.1.2 建立變數
  
  在python中，我們使用`=`來建立變數。舉例來說，`a = b`，代表著我們建立了一個物件a，而因為它是可變的，因此稱為變數，而我們將b這個值指派給a。
  
  在python中，建立變數時不用列出他的資料型態，python會依據b的資料型態來決定a的資料型態。舉例來說，`flt = 1.2`，代表我們建立了一個物件叫做flt，並指定他的值為1.2，而他的資料型態是甚麼呢?因為1.2的資料型態是<浮點數float>，所以從這個例子當中，我們便可以知道，flt的資料型態是float，而他的值為1.2。
  
  從我們出生到現在，我們在數學中學到的`=`代表著等於，然而，在程式語言中，包括python，以及許多的程式語言，都是用`=`賦予值，這是為甚麼呢?有一個原因是，因為標準的鍵盤中，並沒有左箭頭這種具有邏輯意義的符號，且`=`並不會讓人混淆，因此就這麼使用了
  
  * 1.1.2 print()及type()
  
  在python當中，`print()`是一個常用的方法，`print()`的使用方法為`print(值或變數)` ex:print(4)、print("Hello World")。print()是一個非常好用的方法，不只可以印出我們的結果，在debug中也常常使用。
  
  而`type()`也是一個很好用的方法，當我們使用`type()`時，就可以知道值或物件的資料型態。使用方法為:`type(值或物件)` ex:type(4) 
  
  * 1.1.3 變數的命名規則
  
  變數名稱只能用小寫字母、大寫字母、數字、底線，而開頭不能用數字，保留字也不能作為變數的名稱。

-使用\來做一些事

在字串當中，只要打\，後面再加一個字元，就會有特殊的效果 (是"\"不是"/"ㄛ)

比較常用的有 \n  \t  

\n 代表換行的意思

ex:

```

###result = Hello

```

\t 代表tab鍵的意思，移動四格

ex:

```

###result = "Hello    World"
```

-使用[]選字

在字串當中，用[]，加上索引號，就可以取字，索引號從0開始，而如果索引值是負號的話，則會倒過來數

ex:

```
demo = "猜猜我是第幾個字串"
索引值   0 1 2.........8
        -9..........-2-1
demo[0]    ###result : 猜
demo[-1]   ###result : 串
demo[5]    ###result : 幾
demo[9]    ###result : Error 超過範圍了
demo[-5]   ###result : 第
```

-切片處理(slice)

使用[]，裡面有三種參數，開始值、結束值、間隔值。不一定全部的參數都需要，但是如果有間隔值的話，就要有開始值跟結束值

[:] 代表擷取整個字串

[開始值:] 代表擷取開始值至結束的字串

[:結束值] 代表從頭擷取至結束值前一位的字串

[開始值:結束值] 代表擷取從開始值至結束值前一位的字串

[開始值:結束值:間隔值] 代表以間隔值擷取從開始值至結束值前一位的字串

ex:

```
!!!索引值從第0開始

demo = "一二三四五六七八九十"
索引值   0 1.............9
       -10-9............-1
       
demo[:]     ###result : '一二三四五六七八九十' (取全部)
demo[5:]    ###result : '六七八九十'          (從第5個取到最後)
demo[3:8]   ###result : '四五六七八"          (從第3個取到第8個前一位，也就是3取到7)
demo[:9]    ###result : '一二三四五六七八九'   (從頭取到第9個前一位，也就是從0取到8
demo[-8:-5] ###result : '三四五'              (從倒數第8個取到倒數第5個前一位，也就是取-8~-4 
demo[1:5:2] ###result : '二四'                (從第1個取到第5個前一位，隔2取，也就是1到4，中間間隔2，所以取第1，+2後，第3，在加2就爆ㄌ，所以取1  3)
demo[::3]   ###result : '一四七十'            (從頭到尾中，隔3取，所以取第0、3、6、9)
demo[6::2]  ###result : '七九'                (從第6取到尾，隔2取，所以取第6、8)
```

* 大小寫 upper() lower() capitalize() title()

-字串.upper() 可以將字串改成全部大寫

-字串.lower() 可以將字串全部改成小寫

-字串.capitalize() 可以將第一個字母改成大寫

-字串.title() 可以將所有單字的開頭改成大寫

ex:

```

```

* 取代 replace()

-字串.replace(舊字串,新字串,要替換幾個)  如果沒有寫要替換幾個，就只會替換全部

```

| 關係運算子 | python翻中   |

```

### 1.3.2 邏輯運算子

---

| 邏輯運算子 | python翻中   | 
| -------- | -------- | 
|  x and y |  x 且 y  |
|  x or y  |  x 或 y  |
|  not x   |  x 是錯的 |

* x and y

x 和 y都是True則`x and y`為 `True` ，若至少有一個False則 `x and y `為 `False`

* x or y

x 或 y 只要有一個為True，則`x or y`為`True`，若全部都是False則`x or y`為`False`

* not x

若x 為 True，則 `not x`為`False`，若x 為 False，則`not x`為 `True`
